# Pr(score >= x)

## Overview

+ After working through our notebooks, it looks like: neural networks take too long to fit (which especially matters for the subsequent bootstrap), regularized regressions don't predict very well, and random forests and gradient boosting machines are fairly interchangable. 

+ The problem is that not using regressions means we don't get standard errors. If we had standard errors, then we could just look at where a proposed score, X, falls in a normal distribution centered at the predicted score with a standard deviation equal to the model's standard errors. 

+ To get around that issue, we just fit a random forest repeatedly to bootstrapped samples. Then the mean predicted score is our point estimate and the standard deviation is our estimate of the model's standard error. 

+ Once we have those values we can just take a possible total score and calculate 1 - CDF(score X, mean = predicted_score, sd = sd(scores)).

+ The random forests are fit on the past month of games before the game of interest. Having a smaller sample of recent games is faster and tends to be more accurate.

+ Below we demonstrate the technique with a simulation before making predictions on a certain game

Feature engineering -- let's try some non-linear transformations (log, square, cube)

```python
nba_trans['AVGPOINTS_3GAMES_y_sq'] = np.square(nba['AVGPOINTS_3GAMES_y'])
nba_trans['AVGPOINTS_3GAMES_y_cube'] = np.power(nba['AVGPOINTS_3GAMES_y'], 3)
nba_trans['AVGPOINTS_3GAMES_y_log'] = np.log10(nba['AVGPOINTS_3GAMES_y'])
nba_trans['AVGPOINTS_4GAMES_y_sq'] = np.square(nba['AVGPOINTS_4GAMES_y'])
nba_trans['AVGPOINTS_4GAMES_y_cube'] = np.power(nba['AVGPOINTS_4GAMES_y'], 3)
nba_trans['AVGPOINTS_4GAMES_y_log'] = np.log10(nba['AVGPOINTS_4GAMES_y'])
nba_trans['AVGPOINTS_2GAMES_y_sq'] = np.square(nba['AVGPOINTS_2GAMES_y'])
nba_trans['AVGPOINTS_2GAMES_y_cube'] = np.power(nba['AVGPOINTS_2GAMES_y'], 3)
nba_trans['AVGPOINTS_2GAMES_y_log'] = np.log10(nba['AVGPOINTS_2GAMES_y'])
```

```python
nba_trans['pre_PTS_sq'] = np.square(nba['pre_PTS'])
nba_trans['pre_PTS_cube'] = np.power(nba['pre_PTS'], 3)
nba_trans['pre_PTS_log'] = np.log(nba['pre_PTS'])
```

```python
nba_trans['AVGPOINTS_2GAMES_x_log'] = nba_trans['AVGPOINTS_2GAMES_x_log'].replace(-np.inf, np.nan)
nba_trans['AVGPOINTS_3GAMES_x_log'] = nba_trans['AVGPOINTS_3GAMES_x_log'].replace(-np.inf, np.nan)
nba_trans['AVGPOINTS_4GAMES_x_log'] = nba_trans['AVGPOINTS_4GAMES_x_log'].replace(-np.inf, np.nan)
nba_trans['AVGPOINTS_2GAMES_y_log'] = nba_trans['AVGPOINTS_2GAMES_y_log'].replace(-np.inf, np.nan)
nba_trans['AVGPOINTS_3GAMES_y_log'] = nba_trans['AVGPOINTS_3GAMES_y_log'].replace(-np.inf, np.nan)
nba_trans['AVGPOINTS_4GAMES_y_log'] = nba_trans['AVGPOINTS_4GAMES_y_log'].replace(-np.inf, np.nan)
nba_trans['pre_PTS_log'] = nba_trans['pre_PTS_log'].replace(-np.inf, np.nan)
```

## Bootstrap Method Simulation

```python
abs_error = [] # for storing model accuracy
error = []
```

```python
n_iters = 50 # number of games randomly selected to predict
n_samples = 75 # number of samples per bootstrap iteration 
```

### Simulation

This takes about 30min-1hr to run.

```python
random.seed(3210) # for replicability
for i in range(n_iters): # loop through n_iters randomly selected games to predict
    print(i) # Status
    reference_game = nba_trans.sample() # Random game to predict
    nba_no_ref_game = nba_trans[ # all games in last month
        (nba_trans['GAME_DATE'] >= (reference_game.iloc[0,0] - relativedelta(months=+1))) &
        (nba_trans['GAME_DATE'] < reference_game.iloc[0,0])
    ]
    scores = [] # list for storing predicted scores in each boostrap iteration
    for j in range(n_samples): # bootstrap model fitting and then use average as our prediction
        sample_dat = nba_no_ref_game.sample(frac = 0.9, replace = True) # sample with replacement
        X = sample_dat.drop(['TOT_PTS', 'GAME_DATE', 'GAME_ID'], axis = 1) # features
        y = sample_dat['TOT_PTS'] # target
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 925) # train and test
        forest_clf = RandomForestRegressor( # default settings actually work quite well
            random_state = 925, # replicable random forest
        )
        forest_clf.fit(X_train, y_train) # fit predictive model
        # make prediction
        pred_score = forest_clf.predict(reference_game.drop(['TOT_PTS', 'GAME_DATE', 'GAME_ID'], axis = 1))[0]
        scores.append(pred_score) # append prediction to list of predicted scores
    abs_error.append(np.abs(reference_game.iloc[0,15] - np.median(scores))) # median works better w/ outliers
    error.append(reference_game.iloc[0,15] - np.median(scores)) # median works better w/ outliers
```

```python
sns.distplot(error) # error term is roughly distributed around 0 
# on other longer runs this distribution was closer to normal and peaked higher at 0
```

```python
np.mean(error) # on average we are close
```

```python
np.std(error) # standard deviation of error term on this run is about 21 points
# with more iterations this tends to lower
```

So the method works on average. We have somewhat larger tails on the error term than we would like, but it isn't terrible.

## Probability a Game Scores X or More

```python
n_samples = 150 # bootstraps
```

```python
reference_game = nba_trans.sample(random_state = 1234)
```

```python
reference_game # What game are we looking at?
```

```python
# More info on that game:
print(nba[nba['GAME_ID'] == reference_game.iloc[0,1]].GAME_DATE)
print(nba[nba['GAME_ID'] == reference_game.iloc[0,1]].TEAM_ABBREVIATION_x)
print(nba[nba['GAME_ID'] == reference_game.iloc[0,1]].TEAM_ABBREVIATION_y)
```

Training only on past month games:

```python
nba_no_ref_game = nba_trans[ # all games in last month
    (nba_trans['GAME_DATE'] >= (reference_game.iloc[0,0] - relativedelta(months=+1))) &
    (nba_trans['GAME_DATE'] < reference_game.iloc[0,0])
]
```

```python
scores = [] # predicted score for each bootstrap sample
```

```python
for i in range(n_samples):
    print(i) # for status
    sample_dat = nba_no_ref_game.sample(frac = 0.9, replace = True)
    X = sample_dat.drop(['TOT_PTS', 'GAME_DATE', 'GAME_ID'], axis = 1)
    y = sample_dat['TOT_PTS']
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 925)
    forest_clf = RandomForestRegressor( # Default settings work well
        random_state = 925,
    )
    forest_clf.fit(X_train, y_train)
    pred_score = forest_clf.predict(reference_game.drop(['TOT_PTS', 'GAME_DATE', 'GAME_ID'], axis = 1))[0]
    scores.append(pred_score)
```

```python
plt.figure(figsize = (8,5))
fig = sns.distplot(scores)
plt.xlabel("Possible Total Score")
plt.ylabel("Pr(Total Score = X)")
plt.title("Distribution of Possible Total Scores")
plt.savefig("ex_dist.png")
plt.show(fig)
```

```python
np.std(scores) # sd 
```

```python
np.mean(scores) # predicted score
```

```python
reference_game.iloc[0,15] # actual score -- very close! 
```

```python
# prob of greater than or equal to actual score
1 - norm.cdf(x = reference_game.iloc[0,15], loc = np.mean(scores), scale = np.std(scores))
```

### Range of Plausible Scores

```python
score_range = range(
    round(np.mean(scores) - 3*np.std(scores)).astype(int), round(np.mean(scores) + 3*np.std(scores)).astype(int)
)
```

```python
probs = pd.DataFrame() # Where we will store the final prob estimates
```

```python
for i in range(len(score_range)):
    probs = probs.append(pd.DataFrame(
        [score_range[i], (1 - norm.cdf(score_range[i], loc = np.mean(scores), scale = np.std(scores)))]
    ).T)
```

```python
probs.columns = ['score', 'pr_at_least']
```

### Probability of being greater than or equal to a certain score

```python
probs
```

```python
plt.figure(figsize = (8,5))
fig = sns.lineplot(x = 'score', y = 'pr_at_least', data = probs)
plt.xlabel("Possible Total Score")
plt.ylabel("Pr(Total Score >= X)")
plt.title("Chicago at Brooklyn, 2019-02-08 \n Probability Estimates")
plt.savefig("ex_pred.png")
plt.show(fig)
```
